```python
import numpy as np
import matplotlib.pyplot as plt
from collections.abc import Iterable
from matplotlib.patches import Patch  # Per creare elementi di legenda personalizzati
```

```python
def plot_three_boxplots(data1, data2, data3, labels=('A', 'B', 'C'), title='assemblaggio sequenziale'):

    plt.figure(figsize=(7, 5))
    
    # Crea i boxplot
    plt.boxplot([data1, data2, data3],
                labels=labels,
                patch_artist=True,  # riempie le box di colore
                boxprops=dict(facecolor='#b3cde3', color='black'),
                medianprops=dict(color='red', linewidth=2),
                whiskerprops=dict(color='black'),
                capprops=dict(color='black'))
    
    # Titoli e label
    plt.title(title)
    plt.ylabel('microseconds')
    
    plt.show()
```

```python
plot_three_boxplots(dati_assemble, dati_assemble_lambda, dati_assemble_parallel, labels=('seq', 'seq_lambda', 'thredpool_1thread'), title='calcolo triple      nodi: '+str(nodi)+', runs: '+str(len(dati_assemble)))
```
